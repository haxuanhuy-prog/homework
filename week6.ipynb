```python
#w6a9
d1=eval(input())
d2=eval(input())
d={}
for key in d1:
    if key in d2:
        d[key]=d1[key]+d2[key]
    else:
        d[key]=d1[key]
for key in d2:
    if key not in d1:
        d[key]=d2[key]
print(d)
```

```python
#w6a10
nums=list(map(int,input().split()))
k=int(input())
s=[]
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i]+nums[j]==k:
            s.append((nums[i],nums[j]))
s.sort()
print(s)
```
