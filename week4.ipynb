```python
#w4a6
def prime_check(n):
    if n<2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
a,b=map(int,input().split())
prime_sum=0
for i in range(a,b+1):
    if prime_check(i):
        prime_sum+=i
print(prime_sum)
        
```

```python
#w4a10:collatz
def len_collatz(n):
    len=1
    while n!=1:
        if n%2==0:
            n=n//2
        else:
            n=3*n+1
        len+=1
    return len
n=int(input())
lenmax=0
t=0
for i in range(1,n):
    if len_collatz(i)>lenmax:
        lenmax=len_collatz(i)
        t=i
print(t,lenmax)
        
```
