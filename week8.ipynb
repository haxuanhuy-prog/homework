```python
#a1
class cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def S(self):
        return 2*3.14*self.r*(self.h+self.r)
    def V(self):
        return 3.14*(self.r**2)*self.h
r,h=map(float,input().split())
res=cylinder(r,h)
print(f"{res.S():.2f}")
print(f"{res.V():.2f}")
```

```python
#a2
class date:
    def __init__(self,d,m,y):
        self.d=d
        self.m=m
        self.y=y
    def leap_year(self):
        return self.y%4==0 and self.y%100!=0 or self.y%400 ==0
    def months(self):
        if self.m in [1,3,5,7,8,10,12]:
            return 31
        elif self.m in [4,6,9,11]:
            return 30
        elif self.m==2:
            if self.leap_year():
                return 29
            else:
                return 28
    def next_day(self):
        if self.d<self.months():
            self.d+=1
        else:
            self.d=1
            if self.m<12:
                self.m+=1
            else:
                self.m=1
                self.y+=1
    def __str__(self):
        return f"{self.d}/{self.m}/{self.y}"
d,m,y=map(int,input().split("/"))
res=date(d,m,y)
res.next_day()
print(res)
```

```python
#a3
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def goc_toa_do(self):
        return self.x==self.y==0
    def truc_tung(self):
        return self.x==0
    def truc_hoanh(self):
        return self.y==0
    def dist(self):
        return (self.x**2+self.y**2)**0.5
x,y=map(float,input().split())
res=Point(x,y)
if res.goc_toa_do():
    print("M la goc toa do")
elif res.truc_tung():
    print("M nam tren truc tung")
elif res.truc_hoanh():
    print("M nam tren truc hoanh")
else:
    print("M ko nam tren truc tung hoac truc hoanh")
print(f"{res.dist():.2f}")
```

```python
#a5
class ShoppingCart:
    def __init__(self):
        self.items=[]
    def add_item(self,name,price):
        self.items.append({'name':name,'price':price})
    def remove_item(self,name):
        for item in self.items:
            if item['name'] == name:
                self.items.remove(item)
                break
        return "da xoa san pham"
    def is_empty(self):
        return len(self.items) == 0
    def total_price(self):
        total=0
        for item in self.items:
            total += item['price']
        return total
    def show_items(self):
        for item in self.items:
            print(f"{item['name']}: {item['price']}")
    def remove_all_items(self):
        self.items.clear()
cart=ShoppingCart()
while True:
    print("2. Them san pham")
    print("3. Xoa san pham")
    print("4. Kiem tra gio hang co rong ko")
    print("5. Tong gia tri gio hang")
    print("6. Hien thi tat ca san pham")
    print("7. Xoa tat ca san pham")
    option=input("Chon tu 2 den 7: ")
    if option=='2':
        name=input("Nhap ten san pham: ")
        price=float(input("Nhap gia san pham: "))
        cart.add_item(name,price)
    elif option=='3':
        name=input("Nhap ten san pham can xoa: ")
        print(cart.remove_item(name))
    elif option=='4':
        if cart.is_empty():
            print("Gio hang rong")
        else:
            print("Gio hang ko rong")
    elif option=='5':
        print(f"Tong gia tri gio hang: {cart.total_price()}")
    elif option=='6':
        cart.show_items()
    elif option=='7':
        cart.remove_all_items()
        print("Da xoa tat ca san pham")
    else:
        print("Invalid")
    stop=input("Ban co muon thoat??")
    if stop=='yes':
        break
```

```python
#a7
import math
class Fraction:
    def __init__(self,num,den):
        if den<0:
            num=-num
            den=-den
        g=math.gcd(num,den)
        self.num=num//g
        self.den=den//g
    def add(self,other):
        num=self.num*other.den + other.num*self.den
        den=self.den*other.den
        return Fraction(num,den)
    def subtract(self,other):
        num=self.num*other.den - other.num*self.den
        den=self.den*other.den
        return Fraction(num,den)
    def multiply(self,other):
        num=self.num*other.num
        den=self.den*other.den
        return Fraction(num,den)
    def divide(self,other):
        num=self.num*other.den
        den=self.den*other.num
        return Fraction(num,den)
    def __str__(self):
        return f"{self.num}/{self.den}"
a,b,op,c,d=input().split()
a,b,c,d=int(a),int(b),int(c),int(d)
f1=Fraction(a,b)
f2=Fraction(c,d)
if op=='+':
    res=f1.add(f2)
if op=="-":
    res=f1.subtract(f2)
if op=='*':
    res=f1.multiply(f2)
if op=='/':
    res=f1.divide(f2)
print(res)
```

```python
#a8
class BankAccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.balance += amount
        return self.balance
    def withdraw(self,amount):
        self.balance -= amount
        return self.balance
    def __str__(self):
        return f'{self.balance:.2f}'
owner,balance=input().split()
balance=float(balance)
account=BankAccount(owner,balance)
n=int(input("nhập số lệnh: "))
for i in range(n):
    cmd_line=input().split()
    cmd=cmd_line[0]
    amount=float(cmd_line[1])
    if cmd=='DEPOSIT':
        account.deposit(amount)
    elif cmd=='WITHDRAW':
        account.withdraw(amount)
print(account)
    
        
    
```

```python
#a9
class Student:
    def __init__(self,name):
        self.name=name
        self.score={}
    def add_score(self,subject,score):
        self.score[subject]=float(score)
    def avg_score(self):
        return sum(self.score.values())/len(self.score)
    def rank(self):
        avg=self.avg_score()
        if avg>=8:
            return "Excellent"
        elif avg>=6.5:
            return "Good"
        elif avg>=5:
            return "Average"
        else:
            return "Poor"
name=input()
res=Student(name)
n=int(input("nhap so mon hoc va diem:"))
for i in range(n):
    subject,score=input().split()
    score=float(score)
    res.add_score(subject,score)
print(f"{name} {res.avg_score():.2f} {res.rank()}")
```

```python
#a10
class Book:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.library = []
    def add(self,book):
        return self.library.append(book)
    def find_count_author(self,author):
        count = 0
        for i in self.library:
            if i.author == author:
                count += 1
        return count
    def find_countyear_year(self,year):
        count = 0 
        for i in self.library:
            if i.year == year:
                count += 1
        return count
lib = Library()
n = int(input())
for _ in range(n):
    line = input().strip()
    if line.startswith("ADD"):
        _, data = line.split(" ", 1)
        title, author, year = data.split(";")
        book = Book(title, author, int(year))
        lib.add(book)
    elif line.startswith("COUNTYEAR"):
        _,year = line.split()
        print(lib.find_countyear_year(int(year)))
    elif line.startswith("COUNT"):
        _,author = line.split()
        print(lib.find_count_author(author))

    
```

```python
#a11
class Complex:
    def __init__(self,rpart1,rpart2,ipart1,ipart2):
        self.rpart1=rpart1
        self.rpart2=rpart2
        self.ipart1=ipart1
        self.ipart2=ipart2
    def int_mul(self):
        real = self.rpart1*self.rpart2 - self.ipart1*self.ipart2
        return real
    def im_mul(self):
        ima=self.rpart1*self.ipart2+self.rpart2*self.ipart1
        return ima
a,b = map(int,input().split())
c,d = map(int,input().split())
res=Complex(a,c,b,d)
r=res.int_mul()
i=res.im_mul()
print(f"{r} {i}")
        
        
```

```python
#a12
import math
def vector(x,y):
    return (y[0]-x[0], y[1]-x[1], y[2]-x[2])
def cross_product(x,y):
    cp=(x[1]*y[2]-x[2]*y[1], x[2]*y[0]-x[0]*y[2], x[0]*y[1]-x[1]*y[0])
    return cp
def dot_product(x,y):
    dp=x[0]*y[0]+x[1]*y[1]+x[2]*y[2]
    return dp
def norm(x):
    return math.sqrt(x[0]**2 + x[1]**2 + x[2]**2)
A=list(map(float,input().split()))
B=list(map(float,input().split()))
C=list(map(float,input().split()))
D=list(map(float,input().split()))
AB=vector(A,B)
BD=vector(B,D)
AC=vector(A,C)
BC=vector(B,C)
u1=cross_product(AB,AC)
u2=cross_product(BD,BC)
dp=dot_product(u1,u2)
norm1=norm(u1)
norm2=norm(u2)
cos_phi=abs(dp)/(norm1*norm2)
phi=math.degrees(math.acos(cos_phi))
print(f"{phi:.2f}")



    
    
```

```python
#a13
import math
class Triangle:
    def __init__(self,edge1,edge2,edge3):
        self.edge1=edge1
        self.edge2=edge2
        self.edge3=edge3
    def get_area(self):
        a=self.edge1
        b=self.edge2
        c=self.edge3     
        if a+b<=c or a+c<=b or b+c<=a:
            return -1
        p=(a+b+c)/2
        return math.sqrt(p*(p-a)*(p-b)*(p-c))
x,y,z=map(int,input().split())
res=Triangle(x,y,z)
S=res.get_area()
print(f"{S:.2f}")




        
    
```
