```python
#a1
def divide(a,b):
    try:
        t=a/b
        return f"{t:.2f}"
    except ZeroDivisionError:
        return "lỗi"
a,b=map(int,input().split())
res=divide(a,b)
print(res)
```

```python
#a3
def happy_number(x):
        seen=set()
        while x!=1 and x not in seen:
            seen.add(x)
            x=sum(int(digit)**2 for digit in str(x))
        return x==1   
try:
    a=int(input())
    if happy_number(a):
        print("yes")
    else:
       print("no")
except ValueError:
        print("vui long nhap vao mot so!")
    
        
```
