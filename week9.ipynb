```python
#a1
def merge_list(arr1,arr2):
    if not arr1:
        return arr2
    if not arr2:
        return arr1
    if arr1[0]<=arr2[0]:
        return [arr1[0]] + merge_list(arr1[1:],arr2)
    else:
        return [arr2[0]] + merge_list(arr1,arr2[1:])
    
arr1=list(map(int,input().split()))
arr2=list(map(int,input().split()))
res=merge_list(arr1,arr2)
print(res)



    
```

```python
#a5
def check(lst1,lst2):
    if len(lst1) > len(lst2):
        return False
    for i in range(len(lst2)-len(lst1)+1):
        if lst2[i:i+len(lst1)]==lst1:
            return True
    return False

l1=list(map(int,input().split()))
l2=list(map(int,input().split()))
res=check(l1,l2)
print(res)
```
