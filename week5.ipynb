```python
#W5A1
def max_of_two(a,b):
    return max(a,b)
a,b=map(int,input().split())
print(max_of_two(a,b))
```

```python
#W5A2
def swap(a,b):
    return b,a
a,b=map(int,input().split())
a,b=swap(a,b)
print(a,b)
```

```python
#w5a3
def prime_check(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
n=int(input())
print(prime_check(n))
```

```python
#w5a4
def perfect_num(n):
    s=0
    for i in range(1,n):
        if n%i==0:
            s+=i
    return s==n
n=int(input())
print(perfect_num(n))
```

```python
#w5a5
def index_of_k(arr,k):
    for i in range(len(arr)):
        if arr[i]==k:
            return i+1
    return -1
arr=list(map(int,input().split()))
k=int(input())
print(index_of_k(arr,k))
```

```python
#w5a6
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
n=int(input())  
print(factorial(n))
```

```python
#w5a7
def calculate(a,b,o):
    if o=='+':
        return a+b
    elif o=='-':
        return a-b
    elif o=='*':
        return a*b
    elif o=='/':    
        return a/b if b!=0 else "Math Error"
    else:
        return "Syntax Error"
o=input()
for k in ['+','-','*','/']:
    if k in o:
        a,b=map(int,o.split(k))
        print(f"{calculate(a,b,k):.2f}")
        break
```

```python
#w5a8
def hamming(a,b):
    n=a^b
    count=0
    while n>0:
        if n%2!=0:
            count+=1
        n//=2
    return count
a,b=map(int,input().split())
print(hamming(a,b))
```

```python
#w5a9
def sum(n):
    n=str(n)
    s=0
    for i in n:
        s+=int(i)
    return s
n=int(input())
print(sum(n))
```

```python
#w5a10
def dang_cau(a,b):
    map_ab={}
    map_ba={}
    for i in range(len(a)):
        if a[i] in map_ab and map_ab[a[i]]!=b[i]:
            return False
        if b[i] in map_ba and map_ba[b[i]]!=a[i]:
            return False
        map_ab[a[i]]=b[i]
        map_ba[b[i]]=a[i]   
    return True
a,b=map(str,input().split())
print(dang_cau(a,b))
```
